```python
import pandas as  pd 
import  matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
dataset =pd.read_csv("iris_Modified.csv")
dataset.head(3)
```

```python
sns.pairplot(data=dataset)
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
wcss =[]
for i in range(2,21):
    km=KMeans( n_clusters=i,init='k-means++')
    km.fit(dataset)
    wcss.append(km.inertia_)
```

```python
plt.figure(figsize=(10,5))
plt.plot([i for i in range(2,21)],wcss,marker="o")

plt.xlabel("No of cluster")
plt.xticks([i for i in range(2,21)])
plt.ylabel("WCSS")
plt.grid(axis="x")
plt.show()
```

```python
from sklearn.metrics import silhouette_score
```

```python
kmn =KMeans(n_clusters=3)
dataset["Predict"] =kmn.fit_predict(dataset)
```

```python
kmn.labels_
```

```python
silhouette_score(dataset,kmn.labels_)
```

```python
sns.pairplot(data=dataset,hue="Predict")
plt.savefig("predict.jpg")
plt.show()
```

```python
ss =[]
no_c =[j for j in range(2,21)] 
for i in range(2,21):
    km1=KMeans(n_clusters=i)
    km1.fit(dataset)
    ss.append(silhouette_score(dataset,km1.labels_))
```

```python
plt.plot(no_c ,ss)
plt.xlabel("No of Clusters")
plt.ylabel("Silhouti score")
plt.xticks(no_c)
plt.grid(axis="x")
plt.show()
```
